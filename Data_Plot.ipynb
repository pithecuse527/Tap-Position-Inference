```python
from sklearn import preprocessing
import pandas as pd
import numpy as np

SENSOR_DATA_FIEL_PATH = './sample_data/sensor_data.csv'
pd.set_option('display.float_format', lambda x: '%.5f' % x)
sensor_data_all = pd.read_csv(SENSOR_DATA_FIEL_PATH)

sensor_data_tapped_zone_df.dropna()

sensor_data_tapped_zone_df = sensor_data_all.drop(['__id__', 'tap_position_x', 'tap_position_y'], axis=1)
sensor_data_tapped_zone_df.head()
```

```python
data_numpy = sensor_data_tapped_zone_df.to_numpy()

z0_ax = [x[0] for x in data_numpy if x[8] == 0]
z0_ay = [x[1] for x in data_numpy if x[8] == 0]
z0_az = [x[2] for x in data_numpy if x[8] == 0]

z2_ax = [x[0] for x in data_numpy if x[8] == 2]
z2_ay = [x[1] for x in data_numpy if x[8] == 2]
z2_az = [x[2] for x in data_numpy if x[8] == 2]

z9_ax = [x[0] for x in data_numpy if x[8] == 9]
z9_ay = [x[1] for x in data_numpy if x[8] == 9]
z9_az = [x[2] for x in data_numpy if x[8] == 9]

z11_ax = [x[0] for x in data_numpy if x[8] == 11]
z11_ay = [x[1] for x in data_numpy if x[8] == 11]
z11_az = [x[2] for x in data_numpy if x[8] == 11]


z0_gx = [x[3] for x in data_numpy if x[8] == 0]
z0_gy = [x[4] for x in data_numpy if x[8] == 0]
z0_gz = [x[5] for x in data_numpy if x[8] == 0]

z2_gx = [x[3] for x in data_numpy if x[8] == 2]
z2_gy = [x[4] for x in data_numpy if x[8] == 2]
z2_gz = [x[5] for x in data_numpy if x[8] == 2]

z9_gx = [x[3] for x in data_numpy if x[8] == 9]
z9_gy = [x[4] for x in data_numpy if x[8] == 9]
z9_gz = [x[5] for x in data_numpy if x[8] == 9]

z11_gx = [x[3] for x in data_numpy if x[8] == 11]
z11_gy = [x[4] for x in data_numpy if x[8] == 11]
z11_gz = [x[5] for x in data_numpy if x[8] == 11]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 7), dpi=100)
ax = fig.gca(projection='3d')
ax.view_init(50, 70)
zone0 = ax.scatter(z0_ax, z0_ay, z0_az, marker='o')
zone2 = ax.scatter(z2_ax, z2_ay, z2_az, marker='^')
zone9 = ax.scatter(z9_ax, z9_ay, z9_az, marker='v')
zone11 = ax.scatter(z11_ax, z11_ay, z11_az, marker='x')

ax.set_xlabel('accelerometer X')
ax.set_ylabel('accelerometer Y')
ax.set_zlabel('accelerometer Z')
plt.legend([zone0, zone2, zone9, zone11], ['Zone0', 'Zone2', 'Zone9', 'Zone11'])
plt.show()
```
